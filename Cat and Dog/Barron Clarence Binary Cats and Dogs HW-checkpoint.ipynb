```python
def read_image(file_path):
  img = cv2.imread(file_path, cv2.IMREAD_COLOR)
  return cv2.resize(img, (ROWS, COLS),interpolation=cv2.INTER_CUBIC)
```

```python
# Your Part I code here. Print the shapes of all X's and Y's
"""
 With the use of the dictionary, we can sort through the image files and convert the images into
 the dimensions already given to us above, we can re use that same loop, but for every image. Afterward, we can
 convert the dictionaries into Numpy Arrays to use them as test data.
"""
X = []
Y = []

for i, image_file in enumerate(train_images):
    img = cv2.imread(image_file, cv2.IMREAD_COLOR)
    img_resized = cv2.resize(img, (ROWS, COLS), interpolation=cv2.INTER_CUBIC)
    X.append(img_resized.reshape((ROWS * COLS * CHANNELS,)))
    if 'cat' in image_file.lower():
        Y.append(1)
    else:
        Y.append(0)


X = np.array(X)
Y = np.array(Y)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=50)

# Has to be Transposed to give correct shape
X_train = X_train.T
X_test = X_test.T

#Has to be reshaped so it can be changed into 1 row and many columns for each set.
Y_train = Y_train.reshape(1, -1)
Y_test = Y_test.reshape(1, -1)
```

```python
def sigmoid(z):
    a = 1 / (1 + np.exp(-z))
    return a
```

```python
m = X_train.shape[1]
print(m)
```

```python
epoch = 2000
alpha = 2.5

nx = 12288

X_train = X_train/255

W = np.zeros((nx,1))
b = 0
J = np.zeros((epoch,1))

for i in range(epoch):
    #Forward Prop
    Z = np.dot(W.T, X_train) + b
    A = sigmoid(Z)
    L = -((Y_train * np.log(A)) + ((1-Y_train) * np.log(1-A)))
    J[i] = (1/m) * np.sum(L)
    
    
    #Backward Prop
    dZ = A-Y_train
    dW = (1/m) * np.dot(X_train,dZ.T)/m
    db = np.sum(dZ)/m
    W = W - (alpha * dW)
    b = b - (alpha* db)
    
    
print(J[-1])
plt.plot(J[10:-1])
plt.xlabel('epoch')
plt.ylabel('J (Cost Function)')
```

```python
m_train = X_train.shape[1]
A_train = np.shape((1,m_train))
Z_train = np.shape((1,m_train))
Y_predict_train = np.zeros((1,m_train))

Z_train = np.dot(W.T, X_train) + b
A_train = sigmoid(Z_train)

for i in range(A_train.shape[1]):
    if A_train[0, i] >= 0.5:
        Y_predict_train[0, i] = 1
    else:
        Y_predict_train[0,i]  = 0


print("train_accuracy: {} %".format(100-np.mean(np.abs(Y_predict_train - Y_train)) * 100))
```

```python
m_test = X_test.shape[1]
A_train = np.shape((1,m_test))
Z_test = np.shape((1,m_test))
Y_predict_test = np.zeros((1,m_test))

X_test = X_test/255

Z_test = np.dot(W.T, X_test) + b
A_test = sigmoid(Z_test)

for i in range(A_test.shape[1]):
    if A_test[0, i] <= 0.5:
        Y_predict_test[0, i] = 0
    else:
        Y_predict_test[0,i]  = 1

print("test_accuracy : {} %".format(100-np.mean(np.abs(Y_predict_test - Y_test)) * 100))
```

```python
# After successsful learn, with the parameters W and b, download any image of cat or dog and test your classifier using code below:
file = "cat photo.jpg"
test_image = read_image(file)
X_img = test_image.reshape(1, ROWS*COLS*CHANNELS).T/255
Z = np.dot(W.T, X_img) + b
A = sigmoid(Z)
if A>.5:
    print('Prediction: CAT with probability {}%'. format(A*100))
else:
    print('Prediction: DOG with probability {}%'. format(100-A*100))
plt.imshow(test_image)
```
